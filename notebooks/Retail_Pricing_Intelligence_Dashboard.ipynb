```python
df_clean = df.copy()
```

```python
df_clean.head()
```

```python
df_clean["size"].value_counts(dropna=False)
```

```python
df_clean["size"] = df_clean["size"].fillna("Unknown")
```

```python
df_clean["size"].isnull().sum()
```

```python
df_clean["customer_rating"].describe()
```

```python
median_rating = df_clean["customer_rating"].median()
df_clean["customer_rating"] = df_clean["customer_rating"].fillna(median_rating)
```

```python
df_clean["customer_rating"].isnull().sum()
```

```python
df_clean["purchase_date"] = pd.to_datetime(df_clean["purchase_date"])
```

```python
df_clean.info()
```

```python
df_clean["discount_amount"] = (
    df_clean["original_price"] - df_clean["current_price"]
)
```

```python
df_clean.head()
```

```python
bins = [0, 1000, 3000, float("inf")]
labels = ["Budget", "Mid-Range", "Premium"]
df_clean["price_band"] = pd.cut(
    df_clean["current_price"],
    bins=bins,
    labels=labels
)
```

```python
df_clean["price_band"].value_counts()
```

```python
df_clean["purchase_month"] = df_clean["purchase_date"].dt.month_name()
```

```python
df_clean[["purchase_date", "purchase_month"]].head()
```

```python
df_clean["return_status"] = df_clean["is_returned"].map({
    True: "Returned",
    False: "Not Returned"
})
```

```python
df_clean["return_status"].value_counts()
```

```python
df_clean.to_csv(
      "../data/cleaned/retail_pricing_cleaned.csv",
    index=False
)
```

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("../data/retail_pricing.db")
```

```python
df_clean.to_sql(
    "retail_pricing",
    conn,
    if_exists="replace",
    index=False
)
```

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

pd.read_sql(query, conn)
```

```python
#How many products are there in the database?
query = """
SELECT COUNT(*) AS Total_Products
From retail_pricing;
"""

pd.read_sql(query,conn)
```

```python
#How many products are there in each category?
query = """
SELECT
    category,
    COUNT(*) AS Total_Products
FROM retail_pricing
GROUP BY category
ORDER BY Total_Products DESC;
"""

pd.read_sql(query, conn)
```

```python
#What is the average selling price of each brand?
query = """
SELECT
brand,
ROUND(AVG(current_price),2) AS Avg_Selling_Price
FROM retail_pricing
GROUP BY brand
ORDER BY Avg_Selling_Price DESC;
"""

pd.read_sql(query, conn)
```

```python
#Which categories receive the highest average discount?

query = """
SELECT
category,
ROUND(AVG(markdown_percentage),2) AS Avg_Discount
FROM retail_pricing
GROUP BY category
ORDER BY Avg_Discount DESC;
""" 

pd.read_sql(query, conn)
```

```python
#Which brands have the highest customer ratings?
query = """
SELECT
brand,
ROUND(AVG(customer_rating),2) AS Avg_Rating
FROM retail_pricing
GROUP BY brand
ORDER BY Avg_Rating DESC;
"""

pd.read_sql(query, conn)
```

```python
# Which categories have the highest inventory?
query = """
SELECT 
category,
SUM(stock_quantity) AS Total_Stock
FROM retail_pricing
GROUP BY category
ORDER BY Total_Stock DESC;
"""

pd.read_sql(query, conn)
```

```python
#What is the return rate for each category?
query = """
SELECT
category,
ROUND(AVG(is_returned) * 100, 2) AS Return_Rate_Percentage
FROM retail_pricing
GROUP BY category
OrDER BY Return_Rate_Percentage DESC;
"""

pd.read_sql(query, conn)
```
